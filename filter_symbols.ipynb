```python
import Data
import csv
import pandas as pd
import numpy as np
import datetime
```

```python
with open('./datafiles/symbols.csv', 'r') as file:
    reader = csv.reader(file)
    symbols = list(reader)[0]
```

```python
error_symbols = []
for sym in symbols:
    try:
        data = Data.Data(sym, '2010-01-01')
    except:
        error_symbols.append(sym)
        print(error_symbols)
```

```python
filtered_symbols = list(set(symbols) - set(error_symbols))
```

```python
with open('./datafiles/filtered_symbols.csv', 'w') as file:
    wr = csv.writer(file)
    wr.writerow(filtered_symbols)
```
